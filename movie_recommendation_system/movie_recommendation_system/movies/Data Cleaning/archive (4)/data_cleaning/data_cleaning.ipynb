```python
movies = pd.read_csv("c:/Users/vamsh/OneDrive/Desktop/Move Rec System-2/Movie-Recommendation-System-2/movie_recommendation_system/movie_recommendation_system/movies/Data Cleaning/archive (4)/movie.csv")
movies.tail(5)
```

```python
movies = movies.iloc[0:4000]
movies.tail(5)
```

```python
rating = pd.read_csv("c:/Users/vamsh/OneDrive/Desktop/Move Rec System-2/Movie-Recommendation-System-2/movie_recommendation_system/movie_recommendation_system/movies/Data Cleaning/archive (4)/rating.csv")
rating.head(5)
```

```python
rating = rating.iloc[0:4000]
rating.tail(5)
```

```python
movies = movies.merge(rating, on="movieId", how="right")
```

```python
movies.head(5)
```

```python
movies.shape
```

```python
tags = pd.read_csv("c:/Users/vamsh/OneDrive/Desktop/Move Rec System-2/Movie-Recommendation-System-2/movie_recommendation_system/movie_recommendation_system/movies/Data Cleaning/archive (4)/tag.csv")
tags.head(5)
```

```python
tags = tags.iloc[0:4000]
tags.tail(5)
```

```python
movies = movies.merge(tags, on="movieId", how="outer")
movies.head(5)
```

```python
movies.shape
```

```python
movies['title'] = movies['title'].drop_duplicates()
```

```python
movies.head(5)
```

```python
movies = movies.dropna(axis=0)
```

```python
movies.head(5)
```

```python
movies = movies.drop(columns={'userId_x', 'timestamp_x', 'timestamp_y', 'userId_y'})
```

```python
movies.reset_index(drop=True, inplace=True)
movies.head(5)
```

```python
movies[['title_name', 'year']] = movies['title'].str.extract(r"([^(]+) \((\d{4})\)")
```

```python
movies['title'] = movies['title'].str.replace(r"\s\(\d{4}\)", "", regex=True)
```

```python
movies.head(5)
```

```python
movies.shape
```

```python
movies.drop(columns={'title_name'}, inplace=True)
```

```python
movies.head(5)
```

```python
genres = movies['genres'].str.split('|', expand=True).stack().unique()
one_hot_genres =  pd.get_dummies(movies['genres'].str.get_dummies('|'))
```

```python
data_with_genres = movies.join(one_hot_genres)
print(data_with_genres)
```

```python
data_with_genres.tail(5)
```

```python
data_with_genres.drop(columns={'genres'}, inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words='english')

tags_tfidf = tfidf_vectorizer.fit_transform(data_with_genres['tag'])
tags_df = pd.DataFrame(tags_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

data_with_tags = pd.concat([data_with_genres, tags_df], axis=1)
data_with_tags.drop(columns={'tag'}, inplace=True)

data_with_tags.to_csv("c:/Users/vamsh/OneDrive/Desktop/Move Rec System-2/Movie-Recommendation-System-2/movie_recommendation_system/movie_recommendation_system/movies/final data/final_data.csv", index=False)

print(data_with_tags.head())
```
